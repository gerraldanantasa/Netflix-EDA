```python
# Check the type distribution (Movies vs TV Shows)
type_count = df['type'].value_counts().reset_index(name='count')
type_count.columns = ['Type', 'Count']

# Bar Chart with Annotations  
plt.figure(figsize=(16, 10))  

# Create the bar plot  
ax = type_count.plot(kind='bar', x='Type', y='Count', color=['skyblue', 'lightgreen'])  

# Annotate the bars with count and percentage  
total = type_count['Count'].sum()  
for i, v in enumerate(type_count['Count']):  
    percentage = (v / total) * 100  
    plt.text(  
        i,   
        v,   
        f'{v:,}\n({percentage:.1f}%)',   
        ha='center',   
        va='bottom',   
        fontweight='bold',   
        fontsize=6.5  
    )  

plt.title('Distribution of Movies vs TV Shows', fontsize=16, fontweight='bold') 

plt.xlabel('Type', fontsize=14)  
plt.ylabel('Number of Titles', fontsize=14)  
plt.xticks(rotation=0, fontsize=12)  
plt.tight_layout()  
plt.show()  
```

```python
# Pie Chart  
plt.figure(figsize=(5, 3))  

# Create pie chart with percentage labels  
plt.pie(  
    type_count['Count'],   
    labels=type_count['Type'],   
    autopct='%1.1f%%',  # Show percentage  
    colors=['skyblue', 'lightgreen'],  
    explode=[0.1, 0],  # Slightly separate the first slice    
    startangle=90  
)  

plt.title('Proportion of Movies vs TV Shows', fontsize=16, fontweight='bold')  
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle  
plt.tight_layout()  
plt.show()  
```

# Movie count Trend Anlysis

Items added overtime

```python
# Convert the 'date_added' column to datetime and extract the year-month
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_month'] = df['date_added'].dt.to_period('M')

# Count titles by type over time
titles_added_over_time = df.groupby(['year_month', 'type']).size().reset_index(name='count')

titles_over_time_pivot = titles_added_over_time.pivot(index='year_month', columns='type', values='count').fillna(0)
```

```python
titles_over_time_pivot
```

```python



plt.figure(figsize=(16, 8))  

ax = titles_over_time_pivot.plot(  
    kind='line',   
    figsize=(16, 8),   
    marker='o',   
    linewidth=2,  
    colormap='Set3'  
)  
def annotate_max_peak(ax, data):
    for column in data.columns:
        # Find the maximum peak
        max_peak = data[column].max()
        max_peak_date = data[column].idxmax()

        # Convert Period to datetime for plotting
        max_peak_datetime = max_peak_date.to_timestamp()

        plt.annotate(
            f'{column}\n{max_peak:.0f} titles\n{max_peak_date}',
            xy=(max_peak_datetime, max_peak),
            xytext=(10, 10),
            textcoords='offset points',
            fontweight='bold',
            fontsize=9,
            arrowprops=dict(
                arrowstyle='->',
                connectionstyle='arc3,rad=0.2',
                color='darkred'
            )
        )


annotate_max_peak(ax, titles_over_time_pivot)  

plt.title('Trend of Titles Added Over Time (Movies vs TV Shows)', fontsize=18, fontweight='bold')  
plt.xlabel('Year-Month', fontsize=14)  
plt.ylabel('Number of Titles Added', fontsize=14)  
plt.xticks(rotation=45, ha='right', fontsize=10)  
plt.grid(True, linestyle='--', alpha=0.7)  
plt.legend(title='Type', title_fontsize=12, fontsize=10, loc='best', bbox_to_anchor=(1.1, 1))  
plt.tight_layout()  
plt.show() 

```

- Having peaks at 2019-2021
- Started uploading consistent ly uploading content from 2015 and picked up the amount at the start of 2016-2017

```python

print("Detailed Analysis of Title Additions:")  

# Total titles added per type  
total_titles_by_type = titles_over_time_pivot.sum()  
print("\nTotal Titles Added:")  
for type_name, total in total_titles_by_type.items():  
    print(f"{type_name}: {total:.0f} titles")  

# Peak months for each type  
print("\nPeak Months for Each Type:")  
for column in titles_over_time_pivot.columns:  
    peak_month = titles_over_time_pivot[column].idxmax()  
    peak_value = titles_over_time_pivot.loc[peak_month, column]  
    print(f"{column}: {peak_value:.0f} titles in {peak_month}")  
```

Seasonal analysis

```python
# Extract month and season from the date_added column
df['month'] = df['date_added'].dt.month
df['season'] = df['month'].apply(
    lambda x: 'Winter' if x in [12, 1, 2] else
              'Spring' if x in [3, 4, 5] else
              'Summer' if x in [6, 7, 8] else
              'Fall'
)
```

```python
# Count the number of titles added by season and type
season_type_count = df.groupby(['season', 'type']).size().reset_index(name='count')

season_type_pivot = season_type_count.pivot(index='season', columns='type', values='count').fillna(0)

```

```python
# Plot seasonal trends with a bar chart for Movies and TV Shows  
plt.figure(figsize=(14, 8))  

ax = season_type_pivot.plot(  
    kind='bar',   
    stacked=True,   
    color=['lightblue', 'lightgreen']  
)  


season_totals = season_type_pivot.sum(axis=1)  

for i, (movie_count, tv_show_count) in enumerate(season_type_pivot.itertuples(index=False)):  
    # Movies (bottom segment)  
    plt.text(  
        i,   
        movie_count/2,   
        f'Movies\n{movie_count:.0f}',   
        ha='center',   
        va='center',   
        fontsize=9,  
    )  
    
    # TV Shows (top segment)  
    plt.text(  
        i,   
        movie_count + tv_show_count/2,   
        f'TV Shows\n{tv_show_count:.0f}',   
        ha='center',   
        va='center',   
        fontsize=9, 
    )  

plt.title('Seasonal Trends of Titles Added (Movie vs TV Show)', fontsize=18, fontweight='bold')  
plt.xlabel('Season', fontsize=14)  
plt.ylabel('Number of Titles Added', fontsize=14)  
plt.xticks(rotation=0, fontsize=12)  
plt.legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1) 
plt.tight_layout()  
plt.show()  
```

```python
print("Seasonal Trends Detailed Analysis:")  

# Total titles per season  
print("\nTotal Titles per Season:")  
for season, total in season_totals.items():  
    percentage = (total / season_totals.sum()) * 100  
    print(f"{season}: {total:.0f} titles ({percentage:.1f}%)")  

# Breakdown by content type  
print("\nContent Type Breakdown per Season:")  
for season in season_type_pivot.index:  
    print(f"\n{season}:")  
    for content_type, count in season_type_pivot.loc[season].items():  
        percentage = (count / season_type_pivot.loc[season].sum()) * 100  
        print(f"  {content_type}: {count:.0f} titles ({percentage:.1f}%)")  
```

```python
# Prepare the plot
plt.figure(figsize=(16, 10))

# Plot lines for top 10 countries
ax = country_year_pivot.loc[:, country_year_sorted.index].plot(
    kind='line', 
    marker='o', 
    figsize=(16, 10), 
    lw=2,
    colormap='Set3'
)

top_3_countries = country_year_sorted.head(3).index.tolist()

for country in top_3_countries:
    peak_year = country_year_pivot[country].idxmax()
    peak_value = country_year_pivot.loc[peak_year, country]
    plt.annotate(
        f'{country} Peak: {peak_value:.0f} in {peak_year}',
        xy=(peak_year, peak_value),
        xytext=(10, 10),
        textcoords='offset points',
        fontweight='bold',
        color='darkred',
        arrowprops=dict(arrowstyle='->', color='darkred')
    )

plt.title("Trend Analysis of Movies and TV Shows by Country Over Time", fontsize=18, fontweight='bold')
plt.ylabel("Count of Titles", fontsize=14)
plt.xlabel("Year", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Country', title_fontsize=14, fontsize=12,loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Top 3 Countries Peak Years:")
for country in top_3_countries:
    peak_year = country_year_pivot[country].idxmax()
    peak_value = country_year_pivot.loc[peak_year, country]
    print(f"{country}: Peak of {peak_value:.0f} titles in {peak_year}")

```
